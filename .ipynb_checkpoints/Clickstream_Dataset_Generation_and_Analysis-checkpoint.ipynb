<b>Compute time gap between consecutive click events performed by all users</b>

```python
session_id_list = []
event_sequence_list = []
for index in df.index:
    print("index :", index)
    session_id = df['user_ids'][index]
    if session_id not in session_id_list:
        session_id_list.append(session_id)
        print("Creating set S",session_id,sep="")
        event_list = [df['user_click_events'][index]]
        print("Adding set S",session_id," to event_sequence_list", sep="")
        event_sequence_list.append(event_list)
        print(event_sequence_list)
        prev_index = index
    else:
        print("Adding timegap and click event")
        print("index:", index, "prev_index:", prev_index)
        print("df['user_click_event_timestamps'][index]")
        print(df['user_click_event_timestamps'][index])
        print("df['user_click_event_timestamps'][prev_index]")
        print(df['user_click_event_timestamps'][prev_index])
        timegap = df['user_click_event_timestamps'][index]-df['user_click_event_timestamps'][prev_index]
        print("Time gap:")
        print(timegap)
        df.at[index, 'time_gap'] = timegap
        event_sequence_list[-1].append(timegap)
        event_sequence_list[-1].append(df['user_click_events'][index])
        print(event_sequence_list)
        prev_index = index
    print()
```

<b>Convert time gap into seconds</b>

```python
df['time_gap (in seconds)'] = df['time_gap'].apply(lambda delta: int(delta.total_seconds()) if delta else int(0))
df
```

<b>Put time gap into time gap buckets</b>

```python
from datetime import timedelta
# Define the bins for the time gap
# Adjust bin sizes and labels as needed
bins = [timedelta(seconds=0), timedelta(minutes=1), timedelta(hours=1), timedelta(days=1), timedelta(days=365)]
bins_in_seconds = [delta.total_seconds() for delta in bins]
labels = ['<1 min', '1 min - 1 hour', '1 hour - 1 day', '>1 day']
# Put the time gap into buckets using cut()
df['TimeGapBucket'] = pd.cut(df['time_gap (in seconds)'], bins=bins_in_seconds, labels=labels, right=False)
df
```

<b>Map time gap bucket into discrete events</b>

```python
time_gap_event = ['g1', 'g2', 'g3', 'g4']
df['time_gap_events'] = pd.cut(df['time_gap (in seconds)'], bins=bins_in_seconds, labels=time_gap_event, right=False)
df
```

<b>Generate clickstream sequence for each user</b>

```python
def sequence_generator(df):
    session_id_list = []
    event_sequence_list = []
    for index in df.index:
        print("index :", index)
        session_id = df['user_ids'][index]
        if session_id not in session_id_list:
            print("Adding session id:",session_id)
            session_id_list.append(session_id)
            print("Creating set S",session_id,sep="")
            event_list = [df['user_click_events'][index]]
            print("Adding set S",session_id," to event_sequence_list", sep="")
            event_sequence_list.append(event_list)
            print(event_sequence_list)
            prev_index = index
        else:
            print("Adding timegap and click event")
            print("index:", index, "prev_index:", prev_index)
            print("df['user_click_event_timestamps'][index]")
            print(df['user_click_event_timestamps'][index])
            print("df['user_click_event_timestamps'][prev_index]")
            print(df['user_click_event_timestamps'][prev_index])
            time_gap_event = df['time_gap_events'][index]
            event_sequence_list[-1].append(time_gap_event)
            event_sequence_list[-1].append(df['user_click_events'][index])
            print(event_sequence_list)
            prev_index = index
        print()
    return event_sequence_list
```

<b>Prepare dataset for creating similarity graph</b>

```python
from collections import Counter
def computeFrequency(kgram_list):
    kgram_dict = Counter(kgram_list)
    return kgram_dict
```

```python
# n is defined as |T| in the paper
def computeNormalizedFrequency(kgram_dict, n):
    kgram_normalized_dict = {kgram:count/n for kgram, count in kgram_dict.items()}
    return kgram_normalized_dict
```

```python
def representAsFeatureVector(featureWithValues, T):
    featureVector = np.zeros(len(T))
    for index,event in enumerate(T):
        if event in featureWithValues:
            featureVector[index] = featureWithValues[event]
        else:
            featureVector[index] = 0
    return featureVector
```

```python
# Generate similarity score
def computeAngleBetweenTwoVectorsUsingPolarCoordinates(S1_normalized_counts_array, S2_normalized_counts_array):
    print("S1_normalized_counts_array:")
    print(S1_normalized_counts_array)
    print("S2_normalized_counts_array:")
    print(S2_normalized_counts_array)
    # constant
    c = 1/np.pi
    print("constant c:",c)
    # defining the numerator
    nr = np.sum(S1_normalized_counts_array * S2_normalized_counts_array)
    print("numerator value:",nr)
    dr1 = np.power(np.sum(np.power(S1_normalized_counts_array,2)),0.5)
    dr2 = np.power(np.sum(np.power(S2_normalized_counts_array,2)),0.5)
    print("dr1:",dr1)
    print("dr2:",dr2)
    dr = dr1 * dr2
    print("dr:",dr)
    r1 = nr/dr
    print("r1:",r1)
    angle_between_two_vectors_in_radians = np.arccos(r1)
    print("angle_between_two_vectors_in_radians", angle_between_two_vectors_in_radians)
    angle_between_two_vectors_in_degrees = angle_between_two_vectors_in_radians * (180/np.pi)
    print("angle_between_two_vectors_in_degrees", angle_between_two_vectors_in_degrees)
    r = c*angle_between_two_vectors_in_radians
    print("r:", r)
    return r,angle_between_two_vectors_in_degrees
```

<b>Generating 1 gram : k=1</b>

```python
import numpy as np
similarity_graph_df = pd.DataFrame(columns=['node 1', 'node 2', 'Dissimilarity score (k=1)', 'Angle between the two user vectors (degrees)'])
for i in range(no_of_users-1):
    print("user id:",i)
    u1_clickstream_sequence = event_sequence_list[i]
    u1_clickstream_kgram_event_frequency = computeFrequency(u1_clickstream_sequence)
    for j in range(i+1, no_of_users):
        print("user id:",j)
        u2_clickstream_sequence = event_sequence_list[j]
        u2_clickstream_kgram_event_frequency = computeFrequency(u2_clickstream_sequence)
        T = list(set(u1_clickstream_sequence+u2_clickstream_sequence))
        n = len(T)
        u1_clickstream_kgram_event_normalized_frequency = computeNormalizedFrequency(u1_clickstream_kgram_event_frequency, n)
        u2_clickstream_kgram_event_normalized_frequency = computeNormalizedFrequency(u2_clickstream_kgram_event_frequency, n)
        u1_feature_vector = representAsFeatureVector(u1_clickstream_kgram_event_normalized_frequency,T)
        u2_feature_vector = representAsFeatureVector(u2_clickstream_kgram_event_normalized_frequency,T)
        feature_dataset = pd.DataFrame([u1_feature_vector, u2_feature_vector], columns=T)
        print("feature_dataset")
        print(feature_dataset)
        # Lower the score, more similar the 2 user's behavior is
        similarity_score, angle_between_two_vectors_in_degrees = computeAngleBetweenTwoVectorsUsingPolarCoordinates(u1_feature_vector,u2_feature_vector)
        row = ['u'+str(i),'u'+str(j),similarity_score, angle_between_two_vectors_in_degrees]
        similarity_graph_df.loc[len(similarity_graph_df)] = row
        print("similarity_graph:")
        print(similarity_graph_df)
```
